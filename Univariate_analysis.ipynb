```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

**Checking Categorical Data**

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
df['Pclass'].value_counts().plot(kind='bar')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```

**Checking Numerical Data**

1. Histogram: It will give the exact numbers in the range.

```python
df['Age'].plot.hist(bins=10, edgecolor='black')
```

2. Density: It will give the probability of a number to occur.

```python
df['Age'].plot.density()
```

Boxplot: It is a statistical chart that summsrizes the distribution of numerical data.

```python
df['Fare'].plot.box()
```
